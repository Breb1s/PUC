```python
import skimage as sk
from skimage import io
import matplotlib.pyplot as plt
file_path = '/content/drive/MyDrive/PUC/Christian/Camera.jpg'
I2 = sk.io.imread(file_path) # read image using imread function
I2.shape
```

```python
fig = plt.figure(figsize=(10,6))
sk.io.imshow(I2) #Â visualize image using imshow function
sk.io.show()
```

```python
IG = sk.color.rgb2gray(I2)
IG.shape
```

```python
fig = plt.figure(figsize=(10,6))
sk.io.imshow(IG) #Â visualize image using imshow function
sk.io.show()
```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots(1,2,figsize=(12,7))
ax = axes.ravel() #Â convert axes to a 1d array
im1 = ax[0].imshow(IG,cmap=plt.cm.gray) #Â visualize image using imshow function using a grayscale colormap
ax[0].set_title('grayscale image with grayscale colormap') # set figure title
fig.colorbar(im1,ax=ax[0],orientation='horizontal') #Â visualize colorbar
im2 = ax[1].imshow(IG,cmap=plt.cm.inferno) #Â visualize image using imshow function using a grayscale colormap
ax[1].set_title('grayscale image with winter') #Â set figure title
fig.colorbar(im2,ax=ax[1],orientation='horizontal') #Â visualize colorbar
sk.io.show()
```
